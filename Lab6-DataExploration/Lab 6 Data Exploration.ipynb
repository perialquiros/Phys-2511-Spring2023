```python
import matplotlib.pyplot as plt

file = open("Jalen Hurts vs Dak Prescott.csv", mode='r', encoding='utf-8')
data = file.read()
data = data.split(",")
row_len = 17
numOfQBs = 2
for wordID in range(len(data)):
    string = data[wordID]
    newString = string.replace("\n", "")
    newString = newString.replace("\ufeff", "")
    data[wordID] = newString
jhurts_data = data[0:323]
dprescott_data = data[323:]
qbDataSets = [jhurts_data, dprescott_data]
class Stat:
    values = []
    qb = ''
    category = ''
    categoryID = 0
    qbdata = []
    weeksWithData = []
    def __init__(self, categoryID, qbdata):
        self.qbdata = qbdata
        self.categoryID = categoryID
        self.category = qbdata[categoryID]
        self.qb = qbdata[0]
        self.weeksWithData = []
        self.values = []
        self.addValues()
    
    def addValues(self):
        valueID = self.categoryID + row_len
        weekID = 1
        while valueID < len(self.qbdata):
            value = self.qbdata[valueID]
            try:
                value = float(value)
                self.values.append(value)
                self.weeksWithData.append(weekID)
                valueID += row_len
                weekID += 1
            except:
                valueID += row_len
                weekID += 1

jhurts_stats = [Stat(i+3, jhurts_data) for i in range(14)]
dprescott_stats = [Stat(i+3, dprescott_data) for i in range(14)]
dprescott = 0

for categoryID in range(14):
    fig, ax = plt.subplots()
    jhurts_stat = jhurts_stats[categoryID]
    dprescott_stat = dprescott_stats[categoryID]
    jhurts_yvalues = []
    dprescott_yvalues = []
    weeks = range(18)
    category = jhurts_stat.category
    for week in weeks:
        if week in jhurts_stat.weeksWithData:
            weekIndex = jhurts_stat.weeksWithData.index(week)
            jhurts_yvalues.append(jhurts_stat.values[weekIndex])
        else:
            jhurts_yvalues.append(0)
        if week in dprescott_stat.weeksWithData:
            weekIndex = dprescott_stat.weeksWithData.index(week)
            dprescott_yvalues.append(dprescott_stat.values[weekIndex])
        else:
            dprescott_yvalues.append(0)
    ax.plot(weeks, jhurts_yvalues, label='Jalen Hurts', color='green')
    ax.plot(weeks, dprescott_yvalues, label='Dak Prescott', color='blue')
    
    ax.set_xlabel("Game #")
    ax.set_ylabel(category)
    ax.set_xlim(left=0)
    ax.set_ylim(bottom=0)
    plt.legend()
    plt.show()
    plt.close()
    
    print("Comparing Jalen Hurts's and Dak Prescott's " + category + " throughout the regular NFL 2022-2023 season.")
    print("Average (for games played):")
    print("Jalen Hurts: " + str(sum(jhurts_yvalues)/len(jhurts_stat.weeksWithData)))
    print("Dak Prescott: " + str(sum(dprescott_yvalues)/len(dprescott_stat.weeksWithData)))
```
