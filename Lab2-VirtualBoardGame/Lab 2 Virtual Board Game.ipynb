```python
import random as rn
import time as tm
dice = None
turn = 0
playerPositions = []
playerID = 0
gameOver = False
pauseLength = 1
numSpaces = 50
totalPlayers = int(input("How many players? Please input an integer greater than 1. "))
# Define function to move the player
def movePlayer(playerID):
    print("Player " + str(playerID + 1) + "\'s turn! ")
    tm.sleep(pauseLength)
    print("\tRolling dice...")
    tm.sleep(pauseLength)
    spacesMoved = rn.randint(1 , 6)
    potentialPositions = [numSpaces, (playerPositions[playerID] + spacesMoved)]
    playerPositions[playerID] = min(potentialPositions)
    print ("\tYou moved " + str(spacesMoved) + " spaces to land on Space " + str(playerPositions[playerID]))
# Define function to check if the space is a special space
def checkSpace(playerID, sid):
    if sid == 5:
        specialSpace(playerID, sid, 10)
    elif sid == 9:
        specialSpace(playerID, sid, 17)
    elif sid == 12:
        specialSpace(playerID, sid, 4)
    elif sid == 16:
        specialSpace(playerID, sid, 23)
    elif sid == 19:
        specialSpace(playerID, sid, 11)
    elif sid == 22:
        specialSpace(playerID, sid, 43)
    elif sid == 27:
        specialSpace(playerID, sid, 37)
    elif sid == 31:
        specialSpace(playerID, sid, 7)
    elif sid == 36:
        specialSpace(playerID, sid, 44)
    elif sid == 40:
        specialSpace(playerID, sid, 41)
    elif sid == 42:
        specialSpace(playerID, sid, 32)
    elif sid == 47:
        specialSpace(playerID, sid, 48)
    elif sid == 49:
        specialSpace(playerID, sid, 0)
# Define function to add the effect of the special space
def specialSpace (playerID, sid, destination):
    tm.sleep(pauseLength)
    print("\tOh! You landed on Special Space #" + str(sid) + ". You're now on Space #" + str(destination))
    playerPositions[playerID] = destination
# Add player position values at 0 into the playerPositions list.
while len(playerPositions) != totalPlayers:
    playerPositions.append(0)
while gameOver == False:
    movePlayer(playerID)
    checkSpace(playerID, playerPositions[playerID])
    if playerPositions[playerID] == numSpaces:
        print("Player " + str(playerID + 1) + " wins!")
        gameOver = True
    if (playerID + 1) == len(playerPositions):
        playerID = 0
    else:
        playerID = playerID + 1
    tm.sleep(pauseLength)
```
